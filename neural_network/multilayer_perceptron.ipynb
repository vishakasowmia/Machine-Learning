**Oversampling**

SMOTE (Synthetic Minority Over-sampling Technique) is a commonly used oversampling technique for imbalanced datasets. It generates synthetic samples for the minority class by interpolating between the feature vectors of the minority class samples.

```python
x = df.drop(["class"],axis=1)
y = df["class"]
```

```python
#importing SMOTE
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
#plotting count plot using seaborn
sns.countplot(x = yo, palette='muted')
```

**Inference:**
The sample are equally distributed.

**SPLIT THE DATASET**

```python
from sklearn.model_selection import train_test_split
#split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(xo, yo, test_size=0.2, random_state=42)
```

scale our data with StandardScaler() as MLP (and gradient descent) and single layer preceptron is sensitive to un-normalized features. This helps us to speed up our optimization algorithm (gradient descent) and obtain a more accurate classifier.

```python
# Features before mean normalization
unscaled_features = X_train
# Mean Normalization to have a faster classifier
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#calculate Î¼ & Ïƒ(fit) and apply the transformation(transform)
X_train_array = sc.fit_transform(X_train.values)
# Assign the scaled data to a DataFrame & use the index and columns arguments to keep your original indices and column names:
X_train = pd.DataFrame(X_train_array, index=X_train.index, columns=X_train.columns)
# Center test data with the Î¼ & Ïƒ computed (fitted) on training data
X_test_array = sc.transform(X_test.values)
X_test = pd.DataFrame(X_test_array, index=X_test.index, columns=X_test.columns)
```

**single-layer perceptron**



```python
from sklearn.linear_model import Perceptron
```

```python
from sklearn.metrics import accuracy_score
```

```python
# create a Perceptron object
perceptron = Perceptron(max_iter=1000, tol=1e-3)
```

we have max_iter set to 1000 and tol set to 1e-3. We then train the model on the training set using the fit method of the Perceptron object.

```python
# train the model on the training set
perceptron.fit(X_train, y_train)
```

```python
# make predictions on the testing set
y_pred = perceptron.predict(X_test)
```

```python
# evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
# compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:\n', cm)
```

```python
# compute the classification report
cr = classification_report(y_test, y_pred)
print('Classification Report:\n', cr)
```

**Using Multilayer perceptron classifier**
MLPs are used for a wide range of machine learning tasks, such as classification, regression, and prediction. They are particularly useful for modeling complex non-linear relationships between input variables and output variables. MLPs are also highly flexible and can be used with various types of data, including numerical, categorical, and textual data.

```python
from sklearn.neural_network import MLPClassifier

# Initializing the multilayer perceptron
mlp = MLPClassifier(hidden_layer_sizes=(10,), solver='sgd', learning_rate_init=0.01, max_iter=500)
```

**INFERENCE:**
MLP neural network model with one hidden layer of 10 neurons, uses stochastic gradient descent as the optimization algorithm with an initial learning rate of 0.01, and trains the model for a maximum of 500 iterations.

```python
# Train the model
mlp.fit(X_train, y_train)

# Outputs:
MLPClassifier (activation='relu', alpha=0.0001, batch_size='auto', beta_1=0.9,
beta_2=0.999, early_stopping=False, epsilon=1e-08,
hidden_layer_sizes=10, learning_rate='constant',
learning_rate_init=0.01, max_iter=500, momentum=0.9,
nesterovs_momentum=True, power_t=0.5, random_state=None,
shuffle=True, solver='sgd', tol=0.0001, validation_fraction=0.1,
verbose=False, warm_start=False)
```

```python
# Predictions for X_test are compared with y_test
print(mlp.score(X_test, y_test))
```

**INFERENCE:**
From above we can see that Multilayer perceptron has performed well compared it the single layer preceptron it because of the hidden layer which helps us to bring the non-linearity to the layers makes them to learn fast and classify the classes that is a person has diabetes or not.
