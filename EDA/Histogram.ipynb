```python
# Create a figure with multiple subplots
fig, axes = plt.subplots(nrows=len(diabetes.columns), ncols=1, figsize=(10, 30))

# Iterate through each feature and create a histogram
for i, feature in enumerate(diabetes.columns):
    ax = axes[i]
    diabetes[feature].hist(ax=ax, bins=20)
    ax.set_title(feature)

plt.tight_layout()
plt.show()
```

**INFERENCE:**
1. **Pregnancy:** From this histogram, we can infer that there is an outlier at
     17.5.
2. **Glucose:** There is an outlier at 0, the mean of the feature is 79, and the median is 117.

3. **Blood pressure:** There is an outlier at 0, and a mean of 79.

4. **Skin thickness:** The mean might be around 30.

5. **Insulin:**The mean of the feature is 80.

6. **BMI:** The mean of the feature is around 33 because all the data are surrounded there. The median of the feature is 32.

7. **Diabetes pedigree Function:**The mean of the feature is around 0.5; more data are surrounded there and 0.2.

8. **Age:** The mean of the feature is around 30 because most of the data are surrounded around that range. The maximum value is around 20, and the minimum value is around 75. And the mode of the feature is around 22, there is the highest peak.
And the median of the feature is 29.

9. **Outcome:** Since the data are discrete they which has values of 0 and 1.
