**About the Dataset**

This dataset is orginally from Kaggle  the source is:  https://www.kaggle.com/datasets/yasserh/titanic-dataset/.
It contains information about the passengers and survivors aboard the RMS Titanic, which sank on its maiden voyage on April 15, 1912, after hitting an iceberg.

**Attribute Information:**

**PassengerId:** id of the passenger on borad.

**Survived**: whether passenger on board survived or not where 0 = No, 1 = Yes.

**Pclass**: it discribes about the ticket classes which are 1 is 1st class, 2 is 2nd class and 3 is 3rd class.


**Name**: Name of the passenger on board.

**Sex**: The Gender of the passenger on board.

**Age**: Age in years.

**SibSp**: No.of siblings/spouses on board the Titanic.

**Parch**: No.of parents/children on board.

**Ticket**: The ticket number of the passenger.

**Fare**: Passenger fare

**Cabin**: Each of the cabin number

**Embarked**: Port of Embarkation where C is Cherbourg, Q is Queenstown, S is Southampton

**Handling Age feature**

As there are null values in age features we are handling them before we do further analysis.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=titanic_data['Age'])
plt.xlabel('Age')
plt.ylabel('Distribution')
plt.show()
```

**Inference:** From this box plot we can infer that there are outliers present in the dataset. And the median is around 28.

```python

sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Inference:** Histogram we can infer that highest frequency of age is around the 30 and there are null values in the dataset. It is unimodal distribution with single mode (i.e) peak.

**Median Imputation**

Since there is outlier in this feature we use median value to replace the null values. As median is the middle value is not influenced by the outliers therefore we are using that to impute missing values in the data.

```python
titanic_data['Age'].unique()#printing the unique values in the age features using unique function.
```

**Inference:** Where we can clearly see that there are outliers which are 0.64,0.42,0.83,0.67,0.75,0.92 as the age cannot be in points and null value is also present int the data.

```python
titanic_data['Age'] = titanic_data['Age'].apply(lambda x: 0 if pd.notnull(x) and x < 1 else x)
```

```python
titanic_data['Age'].unique()#printing the unique values in the age features using unique function.
```

**Inference:** the age below 1 is set to 0 as the age cannot be 0.76 etc.

```python
median_titanic = titanic_data['Age'].median()# calculating the median of feature
median_titanic
```

```python
titanic_data['Age'].fillna(median_titanic, inplace=True)# imputing features using median
```

**Inference:** Here the there are values like Port of Embarkation where C is Cherbourg, Q is Queenstown, S is Southampton. And nan is the null values.

```python
# Filter the data for each passenger class
first_class_fare = titanic_data[titanic_data['Pclass'] == 1]['Fare'].mean()# Calculating the mean of the features.
second_class_fare = titanic_data[titanic_data['Pclass'] == 2]['Fare'].mean()
third_class_fare = titanic_data[titanic_data['Pclass'] == 3]['Fare'].mean()
labels = ['First Class', 'Second Class', 'Third Class']
fares = [first_class_fare, second_class_fare, third_class_fare]
# ploting the pie chart.
plt.figure(figsize=(8, 8))
plt.pie(fares, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Average Fare by Passenger Class')
plt.show()
```

**Inference:**  From this, it can be inferred that first-class passengers, on average, pay a significantly higher fare (71.0%) compared to second (17.4%) and third class (11.5%) passengers

```python
# Filter the data for each gender
male_fare = titanic_data[titanic_data['Sex'] == 'male']['Fare'].mean()
female_fare = titanic_data[titanic_data['Sex'] == 'female']['Fare'].mean()

# Create a bar chart
labels = ['Male', 'Female']
fares = [male_fare, female_fare]

plt.figure(figsize=(8, 6))
plt.bar(labels, fares, color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Average Fare')
plt.title('Average Fare by Gender')
plt.show()
```

**Inference:**
 It indicates that on average, females have a higher fare than males.


 There are two bars, one labeled male and the other labeled female. The male bar is blue and reaches up to 20 on the y-axis indicating that the average fare for males is 20. The female bar is pink and extends up to 40 on the y-axis, showing that the average fare for females is 40.

```python
# Set the style of seaborn
sns.set(style="whitegrid")

# Create a kernel density estimate (kde) plot to visualize the distribution of ages for Survived and Not Survived
plt.figure(figsize=(12, 8))
sns.kdeplot(data=titanic_data, x="Age", hue="Survived", fill=True, common_norm=False, palette='Set1')
plt.title('Distribution of Ages for Survived and Not Survived Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_columns = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Calculate the correlation matrix
correlation_matrix = titanic_data[numerical_columns].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**Inference:** Heatmap that uses colors ranging from dark red (indicating strong negative correlations) to dark blue (indicating strong positive correlations).

* From this correlation map we can infer that there is negative relationship between  Pclass and Survived (-0.34), meaning as one increases, the other tends to decrease.

* There is a positive correlation between Fare and Survived (0.26), indicating that higher fare is associated with higher survival rate.
